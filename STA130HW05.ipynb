2.In hypothesis testing, the **null hypothesis** and the **alternative hypothesis** serve complementary roles:

1. **Null Hypothesis (H0)**: This is the hypothesis that there is no effect or no difference, and it serves as the default or **starting assumption**. The null hypothesis is what the researcher tries **to disprove or reject through statistical testing**. It typically posits that any observed differences are **due to random chance** rather than a systematic effect.

2. **Alternative Hypothesis (H1 or Ha)**: This hypothesis is considered if the null hypothesis is rejected. It suggests that there is an effect or a difference, and **it represents what the researcher really wants to prove or establish**.

The actual **intended purpose** of setting up a null hypothesis is to create a benchmark for testing whether the observed data can provide sufficient evidence to support an effect or difference that contradicts this default position.

A **good null hypothesis** meets the following criteria:
- **Testability**: It must be formulated in a way that allows it to be tested using statistical methods.
- **Specificity**: It should clearly define the expected outcome under the assumption that the effect or difference being tested does not exist.
- **Neutrality**: It does not suggest any preference towards outcomes and serves as a neutral ground against which changes are measured.

By setting up a robust null hypothesis, researchers ensure that any rejection of this hypothesis in favor of the alternative is based on clear, empirical evidence, reducing the likelihood of drawing incorrect conclusions from random variations in data.

The sentence highlighted from your video underscores a fundamental aspect of hypothesis testing: **the distinction between the sample statistics and population parameters**. Here's how these concepts interrelate and what the statement means:

- **\( x_i \)'s**: These are individual data points within your sample. For example, each \( x_i \) could represent the salary of a single data scientist in a sample of data scientists.

- **Sample Average (\( \bar{x} \))**: This is the arithmetic mean of the sample data points \( x_i \)'s. It serves as an estimate of the population mean but is calculated from only the sample.

- **Population Average (\( \mu \))**: This is the true mean of the entire population from which the sample is drawn. It is usually unknown and is what hypothesis testing aims to make inferences about.

- **Hypothesized Population Average (\( \mu_0 \))**: This is a specific value of the population average that we assume under the null hypothesis. For instance, in your video, \( \mu_0 \) might be $113,000, representing the assumed true mean salary of data scientists.

The crucial point made in the video is that when we perform hypothesis tests, we are not merely commenting on the sample itself (represented by \( \bar{x} \)). Instead, **we use the sample average \( \bar{x} \) to make inferences about the population average \( \mu \)**. **The goal of hypothesis testing is to determine whether the sample provides sufficient evidence to believe that the population parameter \( \mu \) differs significantly from the hypothesized value \( \mu_0 \)**. Thus, even though our calculations are based on sample data, our conclusions aim to reflect truths about the entire population, not just about the sample we analyzed. This distinction is crucial for understanding the broader applicability of the test results.

When we talk about "imagining a world where the null hypothesis is true" in the context of calculating a p-value, we are essentially setting up a hypothetical baseline or benchmark for our statistical analysis. This hypothetical world serves as a reference point to assess the likelihood of observing our actual data, assuming that the null hypothesis holds.

### Why Do We Do This?

**1. Establishing a Baseline:** By assuming the null hypothesis is true, we create a baseline condition where no effect or difference exists. This allows us to compare the observed data against a scenario of "no change" or "no effect," helping us **evaluate how unusual or extreme the actual data are in this neutral context**.

**2. Sampling Distribution Under the Null Hypothesis:** The real power of this approach lies in its use of the sampling distribution of the test statistic under the null hypothesis. This sampling distribution tells us about the behavior of our test statistic when the null hypothesis is true for a variety of possible sample outcomes.

   - **What is a Sampling Distribution?** This is the probability distribution or spread of a statistic (like a mean, difference between means, proportion, etc.) calculated from many hypothetical samples. It represents what we would expect if we repeatedly sampled from the population under the conditions specified by the null hypothesis.

   - **Role in P-Value Calculation:** The p-value is derived by looking at where the actual test statistic falls within this distribution. If the null hypothesis is true, the sampling distribution shows us how much the test statistic could vary due to random chance alone.

### Example for Clarity
Suppose our null hypothesis states that the average IQ score of a population is 100. We take a sample and calculate the average IQ as 105. To understand how extreme this result is:
- We consider a world where the true population average is indeed 100 (null hypothesis is true).
- We look at the sampling distribution of the average IQ under this assumption. This distribution tells us about the range and frequency of average IQ scores we might see from different random samples if the population average really were 100.
- We then determine where our observed average of 105 falls within this distribution. If it falls far from the center (where most outcomes cluster around 100), it suggests that observing a sample average of 105 is rare when the null is true.

**P-Value Interpretation:** Thus, a p-value is the probability of observing a test statistic as extreme as (or more extreme than) the one observed, under the assumption that the null hypothesis is correct. A very low p-value indicates that, under the null hypothesis, the observed result would be highly unlikely, leading us to question the validity of the null hypothesis.

By framing our analysis this way, we can make a statistically informed decision about whether the observed data fit within the expected framework of "no effect," or if they are so unusual that the null hypothesis might not be true, prompting us to consider the alternative hypothesis. This process is central to hypothesis testing and helps ensure that conclusions are not just due to random fluctuations in the data.

### The Role of the Observed Test Statistic
The test statistic calculated from your data (like the mean difference, proportion, etc.) serves as a specific measure to test the hypothesis. It quantifies how far your observed data deviate from what is expected under the null hypothesis.

### Sampling Distribution Under the Null Hypothesis
This distribution provides a model of what the test statistic looks like when the null hypothesis is true across many hypothetical repeats of the experiment or study. It shows all possible values that the test statistic can take, along with how likely they are, if there were no effect (null is true).

### Calculating the P-value
- **Position in the Distribution:** When you calculate a p-value, you're essentially determining where your observed test statistic falls within this null distribution. The p-value is the probability of observing a test statistic as extreme as, or more extreme than, the one derived from your sample data, assuming that the null hypothesis is accurate.

- **Interpreting Extremity:** A "small" p-value indicates that the observed test statistic falls in the tails of this distribution. The tails represent outcomes that are less likely to occur if the null hypothesis is true. The farther into the tail your statistic falls, the lower the probability of seeing such a result by chance alone, under the null hypothesis conditions.

### Why a Small P-Value Challenges the Null Hypothesis
A small p-value (e.g., less than 0.05) suggests that the observed result (represented by the test statistic) would be very unusual if the null hypothesis were true. It means that:
- **Under the Null:** If the null hypothesis were valid, seeing a test statistic as extreme as the one observed would be rare.
- **Rejecting the Null:** Because such an extreme value is unlikely under the null hypothesis, we gain evidence against the null and might consider it implausible or "ridiculous." The smaller the p-value, the stronger the evidence against the null hypothesis, suggesting that perhaps another explanation (like the one proposed in the alternative hypothesis) might be more appropriate.

In essence, a small p-value raises significant doubts about the validity of the null hypothesis by showing that the observed data do not fit well with what would be expected if the null were true. This discord makes the null hypothesis seem less credible, pushing us towards considering alternative explanations.

Simulate a p-value using a "50/50 coin-flipping" model for the assumption of the null hypothesis that the population of humans don't have left or right head tilt tendencies when kissing, and use the table below to determine the level of evidence we have against H_0.


````python
import numpy as np
np.random.seed(1000)
trials = 124
observed_right_tilts = 80
simulations = 10000
#H_0: p=0.5
results = np.random.binomial(trials, 0.5, size=simulations)
#print(results)
p_value = np.mean(results >= observed_right_tilts) # 81,90,100
p_value
````

Since p value equals to 0.0005, there is very strong evidence against the null

7.Can a smaller p-value definitively prove that the null hypothesis is false? Is it possible to definitively prove that Fido is innocent using a p-value? Is it possible to difinitively prove that Fido is guilty using a p-value? How low or high does a p-value have to be to definitely prove one or the other? Explain this concisely in your own words.


A smaller p-value, while suggestive, **cannot definitively prove that the null hypothesis is false**. Similarly, it cannot conclusively prove innocence or guilt in a hypothetical scenario like the one involving Fido. Here's why:

### Understanding P-values
- **Statistical Inference**: A p-value measures the probability of obtaining an observed result (or more extreme) assuming that the null hypothesis is true. It is essentially a measure of how compatible the observed data are with the assumption made under the null hypothesis.
- **Not Proof**: The p-value does not measure the probability that the null hypothesis is true or false. Rather, it indicates the degree of compatibility of the observed data with the null hypothesis. A small p-value indicates that the observed data are unlikely if the null hypothesis were true, leading to doubts about the null hypothesis's validity.

A smaller p-value only indicates that, assuming H0 (the null hypothesis) is true, it is unlikely for the observed statistics or more extreme values to occur, which leads us to question the validity of H0. However, the p-value cannot be used to measure the probability of H0 being true or false; it merely reflects how compatible the observed data is with H0.

### On Proving Innocence or Guilt
- **Fido's Innocence or Guilt**: In the analogy, if a p-value is very small, it suggests that the observed data (e.g., the mess attributed to Fido) are unlikely under the assumption of Fido's innocence (null hypothesis). However, this does not prove Fido's guilt—instead, it undermines the assumption of innocence under the specific circumstances set by the null hypothesis.
- **Alternative Explanations**: A small p-value does not eliminate other possibilities or explanations (such as someone else causing the mess). It only comments on the likelihood of the observed outcome under the null hypothesis.

### Limitations and Conclusions
- **P-values and Decisions**: While a small p-value can lead to rejecting the null hypothesis, it does not provide proof in the absolute sense. Decisions made on the basis of p-values still carry a risk of error—specifically, the risk of a Type I error, where the null hypothesis is incorrectly rejected.
- **Thresholds**: There is no p-value threshold that can definitively prove the null hypothesis true or false. Common thresholds (like 0.05, 0.01) are arbitrary and represent levels of risk for Type I errors that researchers are willing to accept.

### Summary
In essence, p-values help us assess evidence against the null hypothesis but do not provide absolute proof. Decisions based on p-values, whether in statistics or in hypothetical scenarios involving guilt, must be made understanding that they are probabilistic, not definitive. This means embracing uncertainty and acknowledging that statistical tools like p-values are aids in decision-making, not conclusive proofs.

Overview

A most beloved piece of statistical lore about the (most famous) statistician Ronald Fisher involves cups of tea with milk. Fisher and his friend and colleague, Dr. Muriel Bristol, worked at Cambridge in the 1920s and regularly had tea together. During one of their afternoon tea times, Bristol refused a cup of tea from Fisher because he put milk in first BEFORE pouring in the tea. Bristol said she could taste the difference, and much preferred the taste of tea when the milk was poured in afterward the tea. Fisher didn't think that there could be a difference and proposed a hypothesis test to examine the situation.

Fisher made 8 cups of tea, 4 with milk added in first and 4 with tea added in first, and gave them to Dr. Bristol without her seeing how they were made and she would say if she thought the tea or the milk was poured first. As it turned out, Bristol correctly identified if the tea or milk was poured first for all 8 of the cups. Fisher, being a skeptical statistician wanted to test if this could be happening by chance with Bristol just randomly guessing (or whether there was evidence against an assumption of Bristol just randomly guessing), and subsequently designed a statistical hypothesis test to do so.

- \( H_0: p = 0.5 \)
- The null hypothesis states that the proportion \( p \) of students who can correctly identify whether milk or tea was poured first is 0.5, implying no better than random guessing.

**Informal Null Hypothesis:**
- "the probability of STA 130 student can correctly distinguish the order of pouring is 50%."

**Alternative Hypothesis (H1):**
- \( H_1: p > 0.5 \)
- The alternative hypothesis suggests that students can correctly identify the sequence more frequently than random choice.
- "the probability of STA 130 student can correctly distinguish the order of pouring is more than 50%."

### 2. Quantitative Analysis and visulization

#### Methodology:
We will use a binomial test to determine if the observed proportion of correct answers significantly exceeds what would be expected by chance, assuming a binomial distribution under the null hypothesis.

